```python
# distict values in subcategory items
unique_values = df['Sub_Category'].nunique()
print(f'there are: {unique_values} unique values in the subcategory')
```

```python
# Group by 'Sub_Category' and sum the 'Profit'
item_profit_totals = df.groupby('Sub_Category')['Profit'].sum()

# Select the top 10 items
top_10_items = item_profit_totals.nlargest(10)

# Plot the bar chart
ax = top_10_items.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Sub-Categories by Profit')
plt.xlabel('Sub-Categories')
plt.ylabel('Total Profit')

# Annotate the bars with percentages
for p in ax.patches:
    percentage = '{:.1f}%'.format((p.get_height() / top_10_items.sum()) * 100)
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=8, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

```python
# identify the least profitable items
item_profit_totals = df.groupby('Sub_Category')['Profit'].sum()

# Select the least 10 items
top_10_items = item_profit_totals.nsmallest(5)
top_10_items
```

```python
# Visualize in a graph
lowest_profit_items = item_profit_totals.nsmallest(10)
ax = lowest_profit_items.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Sub-Categories with the Lowest Profits')
plt.xlabel('Sub-Categories')
plt.ylabel('Total Profit')

# Annotate the bars with percentages
for p in ax.patches:
    percentage = '{:.1f}%'.format((p.get_height() / lowest_profit_items.sum()) * 100)
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=8, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```
